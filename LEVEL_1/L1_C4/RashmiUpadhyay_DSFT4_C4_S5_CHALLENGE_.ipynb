```python
# TASK- 2
avg=df[(df.make=='BMW')&(df.year==2010)&(df.condition>4.5)]
av1=avg['sellingprice'].mean()
print("Average selling price of year 2010: ",av1)

avg2=df[(df.make=='BMW')&(df.year==2015)&(df.condition>4.5)]
av2=avg2['sellingprice'].mean()
print("Average selling price of year 2015: ",av2)

prediff = (av2-av1)/av2*100

print("Selling price of BMW cars having a score more than 4.5 in Y2010 is less than by Y2015 is by %f%% "%prediff)

```

```python
# TASK- 3

avg=df[(df.interior=='black')]
print("Average mmr for Black Interior: ",avg['mmr'].mean())
avg2=df[(df.interior=='gray')]
print("Average mmr for Gray Interior: ",avg2['mmr'].mean())
```

```python
# TASK- 5

dat1=df[(df.saledate=="Fri Dec 19 2014 09:00:00 GMT-0800 (PST)")& (df.body=="SUV")]
ab=dat1['body'].count()
print("Sale of SUV on Fri Dec 19 2014 09:00:00 GMT-0800 (PST):",ab)

dat2=df[(df.saledate=="Thu Jan 15 2015 04:30:00 GMT-0800 (PST)")& (df.body=="SUV")]
ac=dat2['body'].count()
print("Sale of SUV on Thu Jan 15 2015 04:30:00 GMT-0800 (PST):",ac)

diff = (ab-ac)/ab*100
print("Sale with cars body SUV decreased by %f%% "%diff)
```
