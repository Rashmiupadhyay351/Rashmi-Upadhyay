**NER(Name entity recognition)**:- naming a person, comapny,place recognition etc.

**Entity Linking**:-disambiguating  the textual entities to unique identifiers in a knowledge base.

**Similarity**:- Comparing words and documents how similar they are

**Text Classification**:- assign a set of predefined categories to open-ended text,kind-of-text from documents, medical studies and files, and all over the web.

**Rule Based Matching**:- compared using regular expressions on raw text, spcy rule based matcher engines helps to find the words and phrasesthat we r looking forand also to the tokens within the documens and their relationships

**Training**:- juat a updation of staistical model or prediction

**Processing a pipeline in Spacy**

The pipeline used by the default model consists of a tagger, a parser and an entity recognizer. Each pipeline component returns the processed document, which is then pass on to the next component.

## tips for Processing


* Process the text as a string using nlp.pipe and buffer them in batches instead of one by one. This is usually much more efficient.
*only apply the component you need. To prevent this, disable keyword argument to disble the components you don't need


```python
import spacy
```

```python
text = ['net income was $9.4 million compared to the prior year of 2.7$ million',
        'revenue exceeds twelve billion dollars with a loss of $1b']
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python

docs = nlp.pipe(text, disable = ['tagger', 'parser'])

for doc in docs:
  for ent in doc.ents:
    print(ent.text, ent.label_)
  print()
```

```python
%%timeit
docs = nlp.pipe(text, disable = ['tagger', 'parser'])

for doc in docs:
  for ent in doc.ents:
    print(ent.text, ent.label_)
  print()
```

```python
%%timeit
docs = nlp.pipe(text)

for doc in docs:
  for ent in doc.ents:
    print(ent.text, ent.label_)
  print()
```

## Hashtags and emoji detection

```python
pos_emoji = ["😀", "😃", "😂", "🤣", "😊", "😍"]  # Positive emoji
neg_emoji = ["😞", "😠", "😩", "😢", "😭", "😒"]  # Negative emoji
```

```python
pos = [[{'ORTH': emoji}] for emoji in pos_emoji]
neg = [[{'ORTH': emoji}] for emoji in neg_emoji]
```

```python
pos
```

```python
neg
```

```python
matcher = Matcher(nlp.vocab)
```

```python
def label_sentiment(matcher, doc, i, matches):
  match_id, start, end = matches[i]
  if doc.vocab.strings[match_id] == 'happy':
    doc.sentiment += .1
  elif doc.vocab.strings[match_id] == 'sad':
    doc.sentiment -= .1
```

```python
matcher.add("happy", pos, on_match=label_sentiment)
```

```python
matcher.add("SAD", neg, on_match=label_sentiment)
```

```python
matcher.add('HASHTAG', [[{'ORTH': '#'}, {'IS_ASCII': True}]])
```

```python
doc = nlp(' Congratulations 😀 guys you are about to complete your level 3😍 farewell😭 #Rashmi')
```

```python
matches = matcher(doc)
```

```python
for match_id, start, end in matches:
  string_id = doc.vocab.strings[match_id]
  span = doc[start: end]
  print(string_id, span.text)
```
