```python
# TASK - 2 
#TASK LIST

#(A)                                        

grouped_r_df=cmp_df.groupby(['ram'])
rams=cmp_df.ram.unique()
Strat_Sample_df=pd.DataFrame()

for r in rams:
    r_data=grouped_r_df.get_group(r)
    Sample_data=r_data.sample(n=51, random_state=1, replace= True)
    Strat_Sample_df=Strat_Sample_df.append(r_data, ignore_index=True)
print(Strat_Sample_df)
    
```

```python
# Central tendency of Sample
# (B)
import statistics as st
data_s=Strat_Sample_df['price'].tolist()
s_mean=st.mean(data_s)
s_median=st.median(data_s)
s_sd = st.stdev(data_s)

print("Sample Mean="+str(s_mean))
print("Sample Median="+str(s_median))
print("Sample Standard Deviation="+str(s_sd))
```

```python
# Central tendency of population 

data_p=cmp_df['price'].tolist()
N=len(data_p)
p_mean=st.mean(data_p)
p_median=st.median(data_p)
p_sd = st.stdev(data_p)

print("Population mean="+str(p_mean))
print("Population median="+str(p_median))
print("Population Standard Deviation="+str(p_sd))
```

```python
lower_p=p_mean-4*p_sd
upper_p=p_mean+4*p_sd
norm_p=np.arange(lower_p, upper_p)
plt.plot(norm_p, norm.pdf(norm_p,p_mean,p_sd))
plt.show()
```

```python
# (C)
lower_p=p_mean-4*p_sd
upper_p=p_mean+4*p_sd
norm_p=np.arange(lower_p, upper_p)
plt.plot(norm_p, norm.pdf(norm_p, p_mean ,p_sd))
plt.show()
```

```python
# (D)
lower_s=s_mean-4*s_sd
upper_s=s_mean+4*s_sd
norm_s=np.arange(lower_s, upper_s)
plt.plot(norm_s, norm.pdf(norm_s, s_mean, s_sd))
plt.show()
```

```python
# Calculating Z-Score
# (E)
price= cmp_df['price']
zsc=[]
for i in price:                                
    zsc.append((i-p_mean)/p_sd)
print("Z- score :\n",zsc)
```

```python
# Calculating probability values of Z-Score
# (F)
from scipy.stats import norm

pr_zscore = norm.cdf(zsc)  
print(pr_zscore)
```

```python
#(G)
range_z=np.arange(zsc[0],zsc[-1],0.01)
plt.plot(range_z,norm.pdf(range_z,st.mean(zsc),st.stdev(zsc)))
plt.title("Price standardized")
plt.xlabel("Price amount")
plt.ylabel("Probability of price")
plt.show()
```
