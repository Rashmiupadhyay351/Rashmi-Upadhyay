### TASK - 1

```python
doc = nlp("Ben slapped the mosquito on his arm quite suddenly.")                                                 
for token in doc:
    print('Text=',token.text,', "\tDEP"=',token.dep_, ', "\tHead text"=',token.head.text, ', "\tHead POS"=',token.head.pos_,', "\tChildren"=',
            [child for child in token.children])                                 
```

```python
#Noun chunks

doc = nlp("Ben slapped the mosquito on his arm quite suddenly.")                           
print('Input Text:\n',doc,'\n')                            
for chunk in doc.noun_chunks:
    print('"Text"=',chunk.text,'\t"Root text"=',chunk.root.text, '\t"Root dep"=',chunk.root.dep_,'\t"Root head text"=',chunk.root.head.text)        # Print chunks
```

### Removing stop words

```python
import spacy
from spacy.lang.en import English
nlp = spacy.load("en_core_web_sm")
stopwords = nlp.Defaults.stop_words
with open("Raw_data_for_analysis.txt") as f:
    text=f.read()
lst = []
for token in text.split():
    if token.lower() not in stopwords:        
        lst.append(token)
    
print("Original text : ",text)
print("Text after removing stopwords")
print(''.join(lst))
```

### Lemmatization

### Part Of Speech (POS) tagging

```python
doc = nlp("en_core_web_sm")
f = open("Raw_data_for_analysis.txt")
contents = f.read()                                                             # Read input data
text = str(contents)                                                            # string type
doc = nlp(text)                                                                 # convert to NLP object
for word in doc:                                                                # For all input words
    print(word.text,'\t',word.pos_,'\t',word.tag_)                             # Apply POS and print output
```

### Named Entity Recognition (NER)

```python
doc = nlp("en_core_web_sm")
f = open("Raw_data_for_analysis.txt")
contents = f.read()                                                             
text = str(contents)                                                           
doc = nlp(text)           
print('Input Text:\n',doc)                                                                     
print('\nNER:')
for ent in doc.ents:
    print(ent.text, '\tstart=',ent.start_char, '\tend=',ent.end_char, '\tLabel=',ent.label_)   
    
```

#### Label of word 'Johnson': PERSON

#### Label of word '1997' : DATE

#### Label of word 'School' : N/A
