```python
#TASK- 3.1.3
import numpy as np
indexes=np.arange(0,len(car_df), step=30)
print(car_df.iloc[indexes])
```

```python
# TASK- 3.1.4

grouped_d_df=car_df.groupby(['Year'])
genre=car_df.Year.unique()
Strat_Sample_df=pd.DataFrame()

for d in genre:
    d_data= grouped_d_df.get_group(d)
    sample_data=d_data.sample (n=5, replace= True, random_state=1)
    Strat_Sample_df=Strat_Sample_df.append(sample_data, ignore_index= True)
print(Strat_Sample_df)
```

```python
# TASK- 3.1.5

grouped_dept_df=car_df.groupby(['Year'])
depts=(car_df.Year)
C_Sample_df=pd.DataFrame()

for dept in depts:
    dept_data=grouped_dept_df.get_group(dept)
    sample_data=dept_data.sample(frac=.2, random_state=1)
    C_Sample_df=C_Sample_df.append(sample_data, ignore_index=True)
    
print(C_Sample_df)    
```
