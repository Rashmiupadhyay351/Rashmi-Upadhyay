```python
mean=[]
median=[]
std=[]
coeff=[]

for col in ['Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Minimum_Turning_Radius']:
    mean.append(df.loc[:,col].mean())
    median.append(df.loc[:,col].median())
    std.append(df.loc[:,col].std())
    coeff.append(df.loc[:,col].std()/df.loc[:,col].mean())
    
col_names=['Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Minimum_Turning_Radius']
row_names=['Mean','Median','Standard Deviation','coefficient of variation']

data=pd.DataFrame([mean,median,std,coeff],columns=col_names)
data.insert(0,"Measures",row_names,True)
data
    
```
