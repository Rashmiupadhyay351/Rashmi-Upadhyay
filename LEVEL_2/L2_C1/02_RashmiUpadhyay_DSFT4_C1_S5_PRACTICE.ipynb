```python
# Import the regular expression module
import re

# Compile the pattern: phone
phone = re.compile('\d{3}\-\d{3}\-\d{4}\-{1}')
# The above regular expression means the following 
# \d{3} : Start with a 3 digit number 
# \-    : A hypen (-) after the 3 digit number 
# \d{3} : Another 3 digit number after the previously found hypen(-)
# \-    : A hypen (-) after the 3 digit number 
# \d{4} : A 4 digit number followed by the previously found hypen(-) 

#“120/456-7890(8)”, 

result = phone.match("120/456-7890(8)")# Check if the pattern matches 
print("For 120/456-7890(8) :", bool(result))
#print(result.group())
```

```python
#“123-456-789”, 
phone = re.compile('\d{3}\-\d{3}\-\d{3}')
result = phone.match("123-456-789")# Check if the pattern matches 
print(bool(result))
print(result.group())
```

```python
#“120-456-7890(8)”. 
phone = re.compile('\d{3}\-\d{3}\-\d{3}\-d{1}')
result = phone.match("123-456-7890(8)")# Check if the pattern matches 
print(bool(result))
#print(result.group())
```

```python
import string
text = open("48_hrs.txt","r")
d= dict()
for line in text:
    line = line.strip()
    line= line.lower()
    line=line.translate(line.maketrans("", "",string.punctuation))
    words = line.split(" ")
    for word in words:
        if word in d:
            d[word] = d[word]+1
        else:
            d[word] = 1
for key in list(d.keys()):
    print(key, ":", d[key])
```
