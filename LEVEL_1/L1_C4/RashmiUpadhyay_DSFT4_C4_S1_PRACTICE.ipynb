```python
#TASK 2 

import numpy as np
HR_sal = np.random.randint(25000,50000,500)
print("HR Employees Salaries:\n",HR_sal)
Finance_sal=HR_sal*1.25
print("Finance Employees Salaries:\n",Finance_sal)
Sales_sal=Finance_sal*0.5
print("Sales Employees Salaries:\n",Sales_sal)
IT_sal=HR_sal+5000
print("IT Employee Salaries:\n",IT_sal)
```

```python
#TASK - 3

import numpy as np
Tot_HR=np.sum(HR_sal)
Tot_Finance=np.sum(Finance_sal)
Tot_IT=sum(IT_sal)
Tot_Sales=sum(Sales_sal)
print("Total HR salaries: ",Tot_HR)
print("Total Finance salaries: ",Tot_Finance)
print("Total IT salaries: ",Tot_IT)
print("Total Sales salaries: ",Tot_Sales)
```

```python
#TASK - 4
sal_HR=np.sort(HR_sal)[-1:-11:-1]
print("Top 10 salary of HR Department:\n",sal_HR)
print("Sum of top 10 salary of HR Department: ",sal_HR.sum())

sal_Finance=np.sort(Finance_sal)[-1:-11:-1]
print("Top 10 salary of Finance Department:\n",sal_Finance)
print("Sum of top 10 salary of Finance Department: ",sal_Finance.sum())

sal_Sales=np.sort(Sales_sal)[-1:-11:-1]
print("Top 10 salary of Sales Department:\n",sal_Sales)
print("Sum of top 10 salary of Sales Department: ",sal_Sales.sum())

sal_IT=np.sort(IT_sal)[-1:-11:-1]
print("Top 10 salary of IT Department:\n",sal_IT)
print("Sum of top 10 salary of IT Department: ",sal_IT.sum())


```

```python
#TASK - 5

salhr=np.count_nonzero(HR_sal>45000)
salfin=np.count_nonzero(Finance_sal>45000)
salsal=np.count_nonzero(Sales_sal>45000)
salit=np.count_nonzero(IT_sal>45000)
totalemp= salhr+salfin+salsal+salit
print("Total number of employees with salary more than 45000 that are eligible for 10% increment: ",totalemp)
```

```python
#TASK - 6

sales=np.where(Sales_sal<20000, Sales_sal*1.20, Sales_sal)
Sales_sal=sales.copy()
print("Sales employees:\n",Sales_sal)
print("Employees with 20% increment:\n", sales)
```
